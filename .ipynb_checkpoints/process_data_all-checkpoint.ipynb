```python
import numpy as np
import pandas as pd
```

```python
### Import data
df = pd.read_csv("data/stem.csv")

### Filter data
df = df[df['EMPLOYED'] == 1]
df.drop(columns=['EMPLOYED', 'CaseID', 'weight', 'RECONA_col', 'RECONB_col', 'RECONC_col'], inplace=True)

### Select which columns to keep
# columns_to_keep = ['TEACHSTEM','WORKTYPE_FINAL', 'Xparent', 'ppagect4', 'EDUC4CAT', 'RACE_col', 
#                    'PPGENDER', 'HH_INCOME_col', 'PPREG4', 'ETHN1', 'IDEO', 'HARASS1', 'HARASS2', 
#                    'HARASS3', 'TECH3', 'TECH6']
columns_to_keep = df.columns
# df = pd.DataFrame(df, columns=columns_to_keep)
```

```python
### Purge empty values and 9s
# df = df.astype(float) # change to float so we can insert nans (mark for removal)
df = df.replace([' ', '', 9],[np.nan, np.nan, 9])
df = df.interpolate(method='pad')
# df
```

```python
newdf_columns = []
for column, any_nan in zip(df.columns, df.isna().any()):
    if any_nan == False:
        newdf_columns.append(column)
newdf_columns
df = df[newdf_columns]
# df
```

```python
### Cast everything as integers (empty --> NaN)
df = df.astype(int)

### Remapping values for columns with valuable adjacency
adj_cols = ['ETHN1', 'HARASS1', 'HARASS2', 'STEMJOBa', 'STEMJOBb', 'STEMJOBc', 'STEMJOBd', 
            'STEMJOBe', 'STEMJOBf', 'STEMJOBg', 'STEMJOBh']

df.replace({'ETHN1':{1:1, 2:-1, 3:0}}, inplace=True)
df.replace({'HARASS1':{1:2, 2:1, 3:0}}, inplace=True)
df.replace({'HARASS2':{1:2, 2:1, 3:0}}, inplace=True)
df.replace({'STEMJOBa':{1:1, 2:-1, 3:0}}, inplace=True)
df.replace({'STEMJOBb':{1:1, 2:-1, 3:0}}, inplace=True)
df.replace({'STEMJOBc':{1:1, 2:-1, 3:0}}, inplace=True)
df.replace({'STEMJOBd':{1:1, 2:-1, 3:0}}, inplace=True)
df.replace({'STEMJOBe':{1:1, 2:-1, 3:0}}, inplace=True)
df.replace({'STEMJOBf':{1:1, 2:-1, 3:0}}, inplace=True)
df.replace({'STEMJOBg':{1:1, 2:-1, 3:0}}, inplace=True)
df.replace({'STEMJOBh':{1:1, 2:-1, 3:0}}, inplace=True)

### Remapping values for label columns
label_cols = ['TECH3', 'TECH6']
df.replace({'TECH3':{1:2, 2:1, 3:0}}, inplace=True)
df.replace({'TECH6':{1:2, 2:1, 3:0}}, inplace=True)

### Creating 1hot encoding for the rest
onehot_cols = [elem for elem in df.columns if elem not in adj_cols+label_cols]
print(onehot_cols)

for col in onehot_cols:
    values = set(df[col])
    for value in values:
        df[f'{col}-{value}'] = [1 if elem == value else 0 for elem in df[col]]

df.drop(columns=onehot_cols, inplace=True)

### Export data to CSV
df.to_csv('data/stem_processed_all.csv', index=False)

df
```

```python
unique, counts = np.unique(df.TECH3, return_counts=True)
print(f'3: {counts}, min accuracy={np.round(1000*max(counts)/sum(counts))/10})')
unique, counts = np.unique(df.TECH6, return_counts=True)
print(f'6: {counts}, min accuracy={np.round(1000*max(counts)/sum(counts))/10})')
```
